```python
# =====================================
# Descriptive Statistics
# YRBS 2007
# NoMilkDrinking vs Height
# =====================================

import pandas as pd

# 讀取資料
df = pd.read_csv("../data/processed/YRBS_Cleaned_and_Recoded.csv")

# 各組描述統計
group_stats = (
    df.groupby("NoMilkDrinking")["HowTallAreYouWithoutShoesInMeters"]
      .describe()
      .round(3)
)

print("=== Descriptive Statistics by Group ===")
display(group_stats)

# 整體描述統計
overall_stats = (
    df["HowTallAreYouWithoutShoesInMeters"]
      .describe()
      .round(3)
)

print("\n=== Overall Height Statistics ===")
display(overall_stats.to_frame())
```

```python
# =====================================
# Descriptive Statistics
# YRBS 2007
# NoMilkDrinking vs Height
# =====================================

import pandas as pd
from IPython.display import display

# 讀取資料
df = pd.read_csv("../data/processed/YRBS_Cleaned_and_Recoded.csv")

# 各組描述統計
group_stats = (
    df.groupby("NoMilkDrinking")["HowTallAreYouWithoutShoesInMeters"]
      .describe()
      .round(3)
)

# 美化表格
styled_group_stats = (
    group_stats.style
    .set_caption("Table 1. Descriptive Statistics of Height by Milk Drinking Frequency")
    .format("{:.3f}")
    .set_table_styles([
        {
            "selector": "caption",
            "props": [
                ("font-size", "16px"),
                ("font-weight", "bold"),
                ("text-align", "center")
            ]
        },
        {
            "selector": "th",
            "props": [
                ("text-align", "center")
            ]
        }
    ])
)

display(styled_group_stats)

# 整體描述統計
overall_stats = (
    df["HowTallAreYouWithoutShoesInMeters"]
      .describe()
      .round(3)
      .to_frame(name="Height")
)

styled_overall_stats = (
    overall_stats.style
    .set_caption("Table 2. Overall Descriptive Statistics of Height")
    .format("{:.3f}")
    .set_table_styles([
        {
            "selector": "caption",
            "props": [
                ("font-size", "16px"),
                ("font-weight", "bold"),
                ("text-align", "center")
            ]
        },
        {
            "selector": "th",
            "props": [
                ("text-align", "center")
            ]
        }
    ])
)

display(styled_overall_stats)
```
