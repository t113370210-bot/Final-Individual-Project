**Final Individual Project | YRBS 2007**

## Research Question

> **Is there a significant difference in mean height among students with different levels of milk drinking frequency?**

### Analysis Variables

| Role | Variable | Definition |
|---|---|---|
| Explanatory | `NoMilkDrinking` | Milk drinking frequency, recoded into three ordered groups |
| Response | `HowTallAreYouWithoutShoesInMeters` | Self-reported height without shoes, measured in meters |
### alpha = 0.05

```python
import pandas as pd
from IPython.display import display, HTML

# load dataset
df = pd.read_csv("../data/processed/YRBS_Cleaned_and_Recoded.csv")
```

```python
# Map numeric codes to readable labels
df["Milk Group"] = df["NoMilkDrinking"].map({
    0: "No milk drinking",
    1: "Occasional milk drinking",
    2: "Frequent milk drinking"
})

# Set order of categories
df["Milk Group"] = pd.Categorical(
    df["Milk Group"],
    categories=[
        "No milk drinking",
        "Occasional milk drinking",
        "Frequent milk drinking"
    ],
    ordered=True
)

# Set number display format
pd.set_option("display.float_format", "{:.3f}".format)

# Group descriptive statistics
table1 = (
    df.groupby("Milk Group", observed=False)["HowTallAreYouWithoutShoesInMeters"]
      .describe()
      .round(3)
      .reset_index()
      .rename(columns={"std": "sd"})
)

# Overall descriptive statistics
overall = (
    df["HowTallAreYouWithoutShoesInMeters"]
      .describe()
      .round(3)
      .to_frame()
      .rename(index={"std": "sd"})
)

# Style function 
def full_left(df):
    return (
        df.style
        .set_properties(**{
            "text-align": "left"
        })
        .set_table_styles([
            {
                "selector": "th",
                "props": [("text-align", "left")]
            },
            {
                "selector": "td",
                "props": [("text-align", "left")]
            }
        ])
    )

# Title for group table
display(HTML("""
<h4 style="
    margin-left:265px;
    font-weight:bold;
    margin-bottom:15px;">
Height by Milk Drinking Group
</h4>
"""))

# Display group table
display(full_left(table1))

# Title for overall table
display(HTML("""
<h4 style="
    margin-left:55px;
    font-weight:bold;
    margin-top:25px;
    margin-bottom:10px;">
Overall Height Statistics
</h4>
"""))

# Display overall table
display(full_left(overall))
```
